## 2-5　正規化

### 2-5-4　関数従属

X→Y

Xの値が決まれば、Yの値が **"ただ1つに"** 決まる。<br>
このとき、YはXに関数従属するという。

### 2-5-7　第二正規形

受注日は受注番号に関数従属する → 受注日は受注番号が決まれば一意になる

これは、受注日の種類（受注番号J001も2019/07/01ではないか　など）で考えるのではなく<br>
列「受注日」の値が1つに決まる（受注日の値の選択肢が1つに絞られる）と考えると良い。<br>
つまり、**その列の中で取りうる値が1つに決まる（絞られる）ということ**である。

例えば、$y=2x+1$に$x=1$を代入して$y=3$になるし、$y=x+2$に$x=1$を代入すれば$y=3$になる、といった具合である。<br>
このように、$y$が1つの値になる（一意になる）ことが、関数従属の前提である。<br>
重要なのは「その列の中で取りうる値（選択肢）が1つになること」なので、他と値が被っているかどうかは関係ない。

$y=2x+1$に$x=1$を代入 → $y=3$：一意〇<br>
$y=2x±1$に$x=1$を代入 → $y=3,1$：一意ではない✕

教科書の内容で確認しておく。

受注番号J001に決める → 受注日は2019/07/01と、取りうる値の選択肢が1つとなる<br>
受注番号J001に決める → 商品の選択肢はまだ3つ（マグカップ、銀スプーン、魚ナイフ）があり、1つに絞られていない

#### 部分関数従属

候補キーの一部｛受注番号｝だけで、Yの値がただ1つ決まる → 部分関数従属<br>
候補キーの全て｛受注番号, 商品コード｝が揃ってはじめて、Yの値がただ1つ決まる → 完全関数従属

候補キーである受注番号・商品名・商品コードと、完全関数従属する数量のみ、元の関係に残す。<br>
それ以外の部分関数従属しているものは、他の関係へと切り離す。

### 2-5-8　第三正規形

X：候補キー<br>
Y：候補キーでない属性（列）<br>
Z：非キー属性


X→Y、Y→Z（ただしY→Xではない）とき、Zは間接的に関数従属している（X→Z）ことを、**推移関数従属**しているという。

第三正規形にするには、推移関数従属する部分を別の関係（テーブル）に分離する。

### 2-5-9　ボイスコッド正規形

「部分関数従属する部分を切り出す」という対象が、候補キーの内部にも及ぶ。<br>

例えば、候補キーが A：｛受注番号, 商品名｝とB：｛受注番号、商品コード｝であるとき<br>
Aが候補キーであれば商品コード、Bが候補キーであれば商品名を対象として、部分関数従属していた場合に切り離される。

### 問2.6

一意になる→1行に絞りこむと考えると良い。<br>
また、受注番号は他と被る可能性もあることが前提であることに注意する。


受注日：受注番号は被る可能性があるので、それだけでは1行に絞ることはできない。<br>
受注金額：受注番号は被る可能性があるので、それだけでは1行に絞ることはできない。<br>
顧客名：顧客コードが決まれば1つに絞られるので、部分関数従属している。

主キーが｛受注番号, 顧客コード｝となっており、受注番号だけでないということは<br>
それだけでは一意にできないという可能性がある。よって、受注番号は被るケースも考えられる。
