## 4.1　カテゴリ変数

```python
import numpy as np
import pandas as pd
import mglearn
import os
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

adultデータセットについて、Logistic回帰を用いて分類タスクを行っていく。

### 4.1.1　ワンホットエンコーディング（ダミー変数）

```python
adult_path = os.path.join(mglearn.datasets.DATA_PATH, 'adult.data')

names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation',
         'relationship', 'race', 'gender', 'catital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']

data = pd.read_csv(adult_path, header=None, index_col=False, names=names)

print(data.shape)
data.head()
```

いくつかのカラムに絞り込む。

```python
data = data[['age', 'workclass', 'education', 'gender', 'hours-per-week', 'occupation', 'income']]

print(data.shape)
data.head()
```

#### 4.1.1.1　文字列で表されているカテゴリデータのチェック

各行に含まれているユニークな値と頻度を表示する。<br>
kaggle等でもよく使うので、しっかり覚えておくこと。

```python
print(data.gender.value_counts())
```

```python
print('Original features:\n', list(data.columns), '\n')

data_dummies = pd.get_dummies(data)
print('Features after get_dummies:\n', list(data_dummies.columns))
```

連続特徴量であるageとhours-per-week以外は、列名からワンホットエンコーディングされていることが分かる。

```python
data_dummies.head()
```

values属性を利用して、Numpy配列に変換する。

```python
features = data_dummies.loc[:, 'age':'occupation_ Transport-moving']

X = features.values
y = data_dummies['income_ >50K'].values

print('X.shape:{} y.shape:{}'.format(X.shape, y.shape))
```

これでデータがsklearnで扱える形になった。

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)

print('Test score: {:.2f}'.format(logreg.score(X_test, y_test)))
```

incomeが50Kより多いかどうかの判別モデルを作成することができた。

また、訓練データとテストデータの特徴量の数を一致させるために<br>
両方のデータが入ったDataFrameをget_dummiesすることが重要である。

### 4.1.2　数値でエンコードされているカテゴリ

get_dummiesは、全ての数値データを連続値と認識してしまうため、数値データをワンホットエンコーディングすることができない。<br>
解決方法としては、sklearnのOneHotEncoderを使用するか、数値をカテゴリ変数（文字データ）に置き換えるが有効である。

```python
demo_df = pd.DataFrame({'Integer Feature' : [0, 1, 2, 1], 'Categorical Feature' : ['socks', 'fox', 'socks', 'box']})

display(demo_df)
```

数値データ（Integer Feature）はワンホットエンコーディングできないことを確認する。

```python
pd.get_dummies(demo_df)
```

Integer Featureもワンホットしたい場合は、columnsパラメータで指定すれば良い。

```python
pd.get_dummies(demo_df, columns =['Integer Feature', 'Categorical Feature'])
```

もしくは文字列データに変換してしまえば、自動でワンホットしてくれる。

```python
demo_df['Integer Feature'] = demo_df['Integer Feature'].astype(str)

pd.get_dummies(demo_df)
```
