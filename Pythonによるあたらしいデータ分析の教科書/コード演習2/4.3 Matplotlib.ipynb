```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['spam', 'ham', 'egg']
width = 0.4

# 数取って0.4足すだけ、ということはちょうどy1にぴったりくっつくように中心を0.4ずらしている
x2 = [num + width for num in x]

ax.bar(x, y1, width = width, tick_label = labels, label = 'y1')
ax.bar(x2, y2, width = width, label = 'y2')

ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['spams', 'hams', 'eggs']

# for文である
y_total = [num1 + num2 for num1, num2 in zip(y1, y2)]

# 先にy_totalをプロットする
# tickのあとには_が入る
ax.bar(x, y_total, tick_label = labels, label = 'y1')

# そのあとに、下にくるグラフをプロットする
ax.bar(x, y2, label = 'y2')

ax.legend()

plt.show()
```

```python
np.random.seed(123)

mu = 100
x0 = np.random.normal(mu, 20, 1000)

x1 = np.random.normal(mu, 15, 1000)
x2 = np.random.normal(mu, 10, 1000)

fig, ax = plt.subplots()

labels = ['x0', 'x1', 'x2']

# 同じ( )内に入れると、となり合うグラフが作成できる
ax.hist((x0, x1, x2), label = labels)

plt.show()
```

```python
fig, ax = plt.subplots()

labels = ['x0', 'x1', 'x2']

# stacked = True で積み上げ型（小さい順に下から配置される）
ax.hist((x0, x1, x2), label = labels, stacked = True)
ax.legend()

plt.show()
```

```python
np.random.seed(123)
x = np.random.randn(1000)

fig, ax = plt.subplots()

counts, edges, patches = ax.hist(x, bins = 25)

x_fit = (edges[:-1] + edges[1:]) / 2

# np.sprt：ルート（平方根）
y = 1000 * np.diff(edges) * np.exp(-x_fit ** 2 / 2) / np.sqrt(2 * np.pi)

ax.plot(x_fit, y)

plt.show()
```

```python
fig, ax = plt.subplots()

# family：フォントの種類
# weight：太さ

# ちなみにlinewidthはグラフの線の太さなので全く違う

ax.set_xlabel('xlabel', family = 'fantasy', size = 20, weight = 'bold')

ax.set_ylabel('ylabel', family = 'cursive', size = 40, weight = 'light')

ax.set_title('graph title', family = 'monospace', size = 25, weight = 'heavy')

ax.legend()

plt.show()
```

```python
fontdict = {
    'family':'fantasy',
    'size':'20',
    'weight':'normal',
}

fig, ax = plt.subplots()

ax.set_xlabel('xlabel', fontdict = fontdict)
ax.set_ylabel('ylabel', fontdict = fontdict)

ax.set_title('graph title', fontdict = fontdict, size = 40)

plt.show()
```
