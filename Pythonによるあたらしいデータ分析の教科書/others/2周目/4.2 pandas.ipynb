```python
# dfにない名前を入れると、新しい行もしくは列が作成できる

df.loc[:, 'date'] = df.loc[:, '日付'].apply(pd.to_datetime)
```

```python
# データの上書き

df.loc[:, '摂取カロリー'] = df.loc[:, '摂取カロリー'].astype(np.float32)
```

```python
df = df.set_index('date')
```

```python
df.head()
```

```python
# デフォルトでは昇順

df.sort_values(by = '歩数')
```

```python
df.sort_values(by = '歩数', ascending = False)
```

```python
#列を削除するときは、列方向が適用される

df.drop('日付', axis = 1)
```

```python
df.tail()
```

```python
# 摂取カロリーの列データを上書き

df.loc[:, '歩数/カロリー'] = df.loc[:, '歩数'] / df.loc[:, '摂取カロリー']
df
```

```python
def exercise_judge(ex):
    if ex <= 3.0:
        return'low'
    elif 3.0 < ex < 6.0:
        return'middle'
    else:
        return'high'
```

```python
df.loc[:, '歩数/カロリー'].apply(exercise_judge)
```

```python
# 新しく「運動指数」の列を作成する

df.loc[:, '運動指数'] = df.loc[:, '歩数/カロリー'].apply(exercise_judge)
df
```
