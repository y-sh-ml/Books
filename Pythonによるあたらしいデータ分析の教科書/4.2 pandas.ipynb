```python
df = pd.read_excel("data/201704health.xlsx")
df
```

```python
# 一度dropすると、df自体も変更される（参照になってしまう）
# そういった場合には、もう一度データを読み込んでいく

# axisの参考サイト：https://qiita.com/UD3/items/65eb9416dab37a5562dc
# ニュアンスとしては、それぞれの行に対して日付ラベルを削除する

df = df.drop("日付", axis=1)
```

```python
df.tail()
```

```python
df.loc[:,"歩数/カロリー"] = df.loc[:,"歩数"] / df.loc[:,"摂取カロリー"]
df
```

```python
# return：出力結果（返り値）の設定を行う、出力はしない

def exercise_judge(ex):
    if ex <= 3.0:
        return"Low"
    elif 3.0 < ex <=6.0:
        return"Mid"
    else:
        return"High"
```

```python
 execise_judge(3)
```

```python
# apply：メソッドを適用する

df.loc[:,"運動指数"] = df.loc[:,"歩数/カロリー"].apply(exercise_judge)
df
```

```python
df.to_pickle("data/df_201704helth.pickle")
```

```python
# pd.get_dummies：該当するかどうかを0か1で付ける

df_moved = pd.get_dummies(df.loc[:,"運動指数"], prefix = "運動")
df_moved
```

```python
df_moved.to_pickle("data/df_201704moved.pickle")
```
