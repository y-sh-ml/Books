## 第2章　抽出

```python
import numpy as np
import pandas as pd
```

### 2-1　データ列指定による抽出

```python
reserve_tb = pd.read_csv('reserve.csv')

print(reserve_tb.dtypes)
print(reserve_tb.shape)
reserve_tb.head()
```

```python
reserve_tb.iloc[:, 0:6].head()
```

```python
reserve_tb[['reserve_id', 'hotel_id', 'customer_id', 'reserve_datetime',
            'checkin_date', 'checkin_time', 'checkout_date']][:5]
```

```python
reserve_tb.loc[:, ['reserve_id', 'hotel_id', 'customer_id', 'reserve_datetime', 
                   'checkin_date', 'checkin_time', 'checkout_date']].head()
```

```python
reserve_tb.head()
```

```python
reserve_tb.drop(['people_num', 'total_price'], axis=1, inplace=True)
reserve_tb.head()
```

### 2-2　条件指定による抽出

```python
reserve_tb[(reserve_tb['checkout_date'] >= '2016-10-13') & (reserve_tb['checkout_date'] <= '2016-10-14')].head()
```

```python
reserve_tb.loc[(reserve_tb['checkout_date'] >= '2016-10-13') & (reserve_tb['checkout_date'] <= '2016-10-14'), :].head()
```

```python
reserve_tb.query('"2016-10-13" <= checkout_date <= "2016-10-14"').head()
```

### 2-3　データ値に基づかないサンプリング

```python
reserve_tb.sample(frac=0.5)[:5]
```

### 2-4　集約IDに基づくサンプリング

```python
# ランダムサンプリング
target = pd.Series(reserve_tb['customer_id'].unique()).sample(frac=0.5)

reserve_tb[reserve_tb['customer_id'].isin(target)].head()
```
