## 2-1　データ列指定による抽出

```python
import pandas as pd
import psycopg2
import config
```

```python
connection = psycopg2.connect(**config.connection_config)
```

```python
query ="""

SELECT
  *
FROM
  reserve_tb
LIMIT 5
;

"""

pd.read_sql(query, con=connection)
```

```python
query ="""

SELECT
  reserve_id AS rsv_time,
  hotel_id,
  customer_id,
  reserve_datetime
  checkin_date, checkin_time, checkout_date
FROM
  reserve_tb
LIMIT 5
;

"""

pd.read_sql(query, con=connection)
```
