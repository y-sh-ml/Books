## 2-2　条件指定による抽出

```python
import pandas as pd
import psycopg2
import config
```

```python
connection = psycopg2.connect(**config.connection_config)
```

```python
query ="""

SELECT
  *
FROM
  reserve_tb
LIMIT 5
;

"""

pd.read_sql(query, con=connection)
```

```python
query ="""

SELECT
  *
FROM
  reserve_tb
WHERE
  checkin_date >= '2016-10-12'
  AND checkin_date <= '2016-10-13'
LIMIT 5
;

"""

pd.read_sql(query, con=connection)
```

```python
query ="""

SELECT
  *
FROM
  reserve_tb
WHERE
  checkin_date BETWEEN '2016-10-12' AND '2016-10-13'
LIMIT 5
;

"""

pd.read_sql(query, con=connection)
```
