## 4-1　マスターテーブルの結合

```python
import numpy as np
import pandas as pd
```

```python
reserve_tb = pd.read_csv('reserve.csv')
print(reserve_tb.shape)
reserve_tb.head()
```

```python
hotel_tb = pd.read_csv('hotel.csv')
print(hotel_tb.shape)
hotel_tb.head()
```

```python
pd.merge(reserve_tb, hotel_tb, on='hotel_id', how='inner').query("people_num == 1 & is_business").head()
```

```python
pd.merge(reserve_tb.query('people_num==1'), hotel_tb.query('is_business'), on='hotel_id', how='inner').head()
```
