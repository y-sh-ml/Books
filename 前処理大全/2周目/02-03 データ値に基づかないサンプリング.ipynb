## 2-3　データ値に基づかないサンプリング

```python
import pandas as pd
```

```python
reserve_tb = pd.read_csv('reserve.csv')
print(reserve_tb.shape)
reserve_tb.head()
```

`frac` で、抽出するデータの割合を指定できる。

```python
reserve_tb.sample(frac=0.5).reset_index(drop=True).head()
```

`n` で、サンプリングする件数を指定できる。

```python
reserve_tb.sample(n=100).reset_index(drop=True).head()
```
