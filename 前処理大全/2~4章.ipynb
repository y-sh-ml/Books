### 4-2　条件に応じた結合テーブルの切り替え

レコメンド：その商品を買ってくれそうな人に勧めること（Amazonのようなイメージ）

```python
hotel_tb = pd.read_csv('hotel.csv')

print(hotel_tb.shape)
```

```python
# 必要のないメモリ解放のためのライブラリ
import gc
```

①閲覧しているホテルと、レコメンド（推薦）するホテルの地域を設定

```python
# small_area_nameごとにホテル数をカウント(cnt)
small_area_mst = hotel_tb.groupby(['big_area_name', 'small_area_name'], as_index=False).size().reset_index()
small_area_mst.columns = ['big_area_name', 'small_area_name', 'hotel_cnt']

small_area_mst.head()

# 情報を把握しやすいように、big_area_nameも入れている
```

join_area_id：レコメンドで表示するホテルの領域（範囲）

```python
# 20件以上であれば、join_area_idにsmall_name_areaを代入（20件以下なら、big_name_areaを代入）
small_area_mst['join_area_id'] = np.where(small_area_mst['hotel_cnt'] -1 >= 20,
                                          small_area_mst['small_area_name'],
                                         small_area_mst['big_area_name'])

# np.where(条件式, x, y)：条件式がTrueならxを、Falseならyを適用する

small_area_mst.head()
```

```python
# 必要なくなった列の削除
small_area_mst.drop(['hotel_cnt', 'big_area_name'], axis=1, inplace=True)
small_area_mst.head()

# join_area_idを追加できたので、2つのカラムは不要
```

```python
hotel_tb.head()
```

```python
# デフォルトはhow='inner'
base_hotel_mst = pd.merge(hotel_tb, small_area_mst, on='small_area_name')

base_hotel_mst.head()
```

```python
base_hotel_mst = pd.merge(hotel_tb, small_area_mst, on='small_area_name').loc[:, ['hotel_id', 'join_area_id']]

base_hotel_mst.head()
```

②レコメンドされる（推薦される）ホテルと、その地域情報をまとめる

```python
del small_area_mst
gc.collect()
```

```python
hotel_tb.head()
```

```python
# small_area_name, big_area_nameをjoin_area_idのカラム名に変更

recommend_hotel_mst = pd.concat([hotel_tb[['small_area_name', 'hotel_id']] \
                                 .rename(columns={'small_area_name' : 'join_area_id'}, inplace=False),
                                 
                                hotel_tb[['big_area_name', 'hotel_id']] \
                                 .rename(columns={'big_area_name' : 'join_area_id'}, inplace=False)], axis=0)

recommend_hotel_mst.head()
```

```python
recommend_hotel_mst.rename(columns={'hotel_id' : 'rec_hotel_id'}, inplace=True)

recommend_hotel_mst.head()
```

③閲覧したホテルとレコメンドで表示するホテルを結びつける（重複は削除する）

```python
recommends = pd.merge(base_hotel_mst, recommend_hotel_mst, on='join_area_id') \
.loc[:, ['hotel_id', 'rec_hotel_id']] \
.query('hotel_id != rec_hotel_id')

print(recommends.shape)
recommends.head()
```

これで、h_1を閲覧したときに、レコメンド（推薦）対象となるホテルがタグ付けされている

#### 過去n日間の平均値

```python
reserve_tb = pd.read_csv('reserve.csv')

print(reserve_tb.shape)
reserve_tb.head()
```

```python
# customer_idごとに、reserve_datetimeで並び替えをする
result = reserve_tb.groupby('customer_id').apply(lambda x: x.sort_values(by='reserve_datetime', ascending=True))\
.reset_index(drop=True)

# ばらばらになったインデックス番号を上から振り直し、前のインデックス番号（customer_id）を削除している
```

reset_index：https://note.nkmk.me/python-pandas-reset-index/

```python
result.head()
```

```python
result['price_avg'] = pd.Series(result.groupby('customer_id')['total_price']\
                                .rolling(center=False, window=3, min_periods=1).mean().reset_index(drop=True))

result.head()
```

自身のデータ行を含めて、設定した件数になるように上のデータを加えている。<br>
例えば3行目の828866は、それに対応している194499と上2つ（20600, 33600）の平均値である。<br>
このように算出した82866を1つ下にずらすことで、自身を含まない上3つのデータの平均値を挿入している。<br>

```python
result['price_avg'] = result.groupby('customer_id')['price_avg'].shift(periods=1)
result.head()
```
