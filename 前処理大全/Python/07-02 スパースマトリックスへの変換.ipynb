```python
import pandas as pd
from scipy.sparse import csc_matrix
```

```python
reserve_tb = pd.read_csv(f'{path}/reserve.csv')
print(reserve_tb.shape)
reserve_tb.head()
```

```python
cnt_tb = reserve_tb.groupby(['customer_id', 'people_num'])['reserve_id'].size().reset_index()
cnt_tb.columns = ['customer_id', 'people_num', 'rsv_cnt']
print(cnt_tb.shape)
cnt_tb.head()
```

```python
customer_id = pd.Categorical(cnt_tb['customer_id'])
people_num = pd.Categorical(cnt_tb['people_num'])

print(customer_id.codes[:4])
print(people_num.codes[:4])

print(customer_id.categories[:5])
print(people_num.categories[:5])
```

```python
df = csc_matrix((cnt_tb['rsv_cnt'], (customer_id.codes, people_num.codes)), shape=(len(customer_id.categories), len(people_num.categories)))
df
```

```python
df = pd.DataFrame(df.toarray())
df.head()
```
