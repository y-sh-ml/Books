```python
import numpy as np
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
production_miss_num.replace('None', np.nan, inplace=True)
production_miss_num.dropna(subset=['thickness'], inplace=True)
```

```python
production_miss_num = pd.read_csv(f'{path}/production_missing_num.csv')
print(production_miss_num.shape)
production_miss_num.head()
```

```python
production_miss_num.replace('None', np.nan, inplace=True)
production_miss_num.fillna({'thickness': 1}, inplace=True)
```

```python
production_miss_num = pd.read_csv(f'{path}/production_missing_num.csv')
print(production_miss_num.shape)
production_miss_num.head()
```

```python
production_miss_num.replace('None', np.nan, inplace=True)
production_miss_num['thickness'] = production_miss_num['thickness'].astype('float64')

thickness_mean = production_miss_num['thickness'].mean()
production_miss_num.fillna({'thickness': thickness_mean}, inplace=True)
```

```python
production_miss_num = pd.read_csv(f'{path}/production_missing_num.csv')
print(production_miss_num.shape)
production_miss_num.head()
```

```python
production_miss_num.replace('None', np.nan, inplace=True)
production_miss_num['thickness'] = production_miss_num['thickness'].astype('float64')
production_miss_num['type'] = production_miss_num['type'].astype('category')
production_miss_num['fault_flg'] = production_miss_num['fault_flg'].astype('category')

production_dummy_flg = pd.get_dummies(production_miss_num[['type', 'fault_flg']], drop_first=True)
```

```python
X = pd.concat([production_miss_num[['length', 'thickness']], production_dummy_flg], axis=1)
print(X.shape)
X.head()
```

```python
imputer = IterativeImputer(
    random_state=0,
    sample_posterior=True
)
```

```python
production_imputed = imputer.fit_transform(X)
production_imputed = pd.DataFrame(production_imputed, columns=X.columns)
```
