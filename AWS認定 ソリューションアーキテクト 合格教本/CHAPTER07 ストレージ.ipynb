# CHAPTER7　ストレージ

## 7-1　S3（Amazon Simple Storage Service）

・S3：AWSが提供する、安価で高い耐久性を持つストレージサービス

S3の主な要素は、以下となる。

・バケット：オブジェクトの**保管場所**<br>
・オブジェクト：**データ本体**（データそのもの）<br>
・キー：オブジェクトの**格納URLパス**、「バケット名＋オブジェクト名＋バージョン」が一意になる

S3の活用方法としては、以下の通りである。

・**OSプロビジョニング用コンフィグの設置**<br>
・各種ログの保管<br>
・バックアップデータの保管

※プロビジョニング：必要に応じてネットワークやコンピューター設備などを提供できるように予測し、準備しておくこと

S3で操作したデータには、以下の性質を保証する。

・データの更新、削除：整合性<br>
・データの追加：一貫性

## 7-2　S3のストレージクラスとGlacierによるコスト最適化

S3のストレージクラスには、以下の種類がある。

・標準：S3の標準となるストレージクラス、取り出し料金なし<br>
・標準 IA：低頻度アクセス向き、**長期保存**に向いている、取り出し料金あり<br>
・1ゾーン IA：低頻度アクセス向き、**セカンダリバックアップ**に向いている、S3 標準クラスのAVZを必要としない、取り出し料金あり<br>
・Glacier：**アクセスがほとんどない時**に向いている、**最も安価に保存**できる、取り出し料金あり

上記に対して、下記のオプションも適用できる。

・ライフサイクルポリシー：設定した日数経過後にストレージクラスを自動的に変更する仕組み

## 7-3　S3のセキュリティ

・**バケットポリシー**：S3バケットポリシーに対するアクセス権限を設定する機能<br>
・**CloudTrail**：S3のすべての操作ログを記録するもの

バケットポリシーは、JSONベースで以下のように定義する。

・どのバケット、もしくはオブジェクトに対して（Resource）<br>
・どのAPI操作を（Action）<br>
・誰から（Principal）<br>
・許可・拒否するのか（Effect）

**S3の暗号化**には2種類ある。

・クライアントサイド暗号化：暗号化や復号化をクライアントサイド（ユーザ側）で行う<br>
・サーバーサイド暗号化：暗号化や復号化をサーバーサイド（AWS側）に任せる

セキュアなファイルへのアクセスは、IAMを使用すると**アクセスキー・シークレットキーが必要**となってしまうため<br>
下記いずれかの方法で行うことが推奨されている。

・IAMロールをEC2に付与してS3にアクセスする<br>
・CloudTrailのS3 Loggingオプションをオンにする

※CloudTrail：S3のすべての操作ログを記録するもの

## 7-4　S3 Web ホスティング

通常、Webサイトの公開には、Webサーバーを準備して構築・運用を行う必要があるが<br>
**S3 Web ホスティング**はそれらの工程を省いて、簡単にWebサイトをホスティングすることができる。

## 7-5　EBS（Elastic Block Store）

・EBS：EC2インスタンスとネットワーク接続して使用するストレージサービス<br>
・インスタンスストア：ホストコンピュータのストレージを使用するストレージサービス

EC2インスタンスの停止時に、以下のような違いがある。

・EBS：データは消失しない<br>
・インスタンスストア：データが消失する（ただしEC2インスタンスの停止ではなく再起動であれば、データは消失しない）

EBSには、以下のボリュームタイプがある。

・EBSマグネティック：**旧世代のHDD**、データへのアクセスが低いワークロードで使用、課金なし<br>
・Cold HDD（sc1）：アクセス頻度の低いワークロード用の**低コストHDD**ボリューム、課金あり<br>
・スループット最適化 HDD（st1）：アクセス頻度とスループットが高いワークロードで**低コストのHDD**ボリューム、課金あり

・汎用SSD（gp2）：**コストパフォーマンスに優れる汎用SSD**ボリューム、ほとんどのワークロードで使用、課金あり<br>
・プロビジョンド IOPS SSD（io 1）：低レイテンシーor高スループットワークロードで**高パフォーマンスのSSD**ボリューム、課金あり<br>

各用語についても、補足しておく。

・ワークロード：クラウドアプリケーションを構成するリソースとコードのコレクション<br>
・スループット：処理能力（単位当たりに処理可能な量）<br>
・レイテンシー：通信の遅延時間

他にも、以下の機能がある。

・EBS スナップショット：EBSのデータを Amazon S3 にバックアップする機能

EBS スナップショットの特徴は、以下の通りである。

・オンラインで取得できる<br>
・ルートボリュームのスナップショットが取得できる<br>
・単一のEC2インスタンスにしかアタッチできない<br>
・スナップショットは開始時のものが保管されるため、その後であればデータを変更したり削除しても問題ない

## 7-6　EFS（Elastic File System）

・EBSは1つのEC2からしかアクセスできないストレージサービス<br>
・EFSは複数のEC2からアクセスできるストレージサービス
