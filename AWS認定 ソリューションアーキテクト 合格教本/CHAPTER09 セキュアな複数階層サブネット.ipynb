# CHAPTER9　セキュアな複数階層サブネット

## 9-1　セキュアな複数階層サブネット

パブリックサブネットとプライベートサブネットを作成し、複数の階層を持つサブネットを構成する。

理由としては、以下の通りである。

・データベースやサーバーにインターネットから直接アクセスすることを防ぐ<br>
・プロキシサーバーやインターネットゲートウェイ経由でしか、インターネットにアクセスができないようにする

## 9-2　階層パターン

**1階層サブネット**の特徴は、以下である。

・サブネットがシンプルである<br>
・すべてのサーバーが、インターネットからアクセス可能なサブネットに配置される

**2階層サブネット**の特徴は、以下である。

・Webサーバーをパブリックサブネット、DBサーバーをプライベートサブネットに配置する<br>
・データベースをインターネットから直接アクセスできないネットワークに配置できる

**3階層サブネット**の特徴は、以下である。

・**ELB**（負荷分散装置）と**Basion**（踏み台サーバー）をパブリックサブネットに配置する<br>
・Webサーバーを**プライベートサブネットA**に、データベースサーバーを**プライベートサブネットB**に配置する<br>
・ロードバランサーとNginx（リバースプロキシ）の代わりに、ELBを使用することも多い

キーワードの補足もしておく。

・プロキシ：「代理」という意味で、**内部ネットワークから**インターネットに接続する際に中継すること<br>
・リバースプロキシ：**外部インターネットから**サーバーへアクセスされる通信を中継すること、プロキシの逆（リバース）である

## 9-3　セキュリティ面でのメリット

・NATゲートウェイ：プライベートサブネットからインターネットへアクセスする際に、**パブリックサブネットに配置**する<br>
・踏み台（Bation）：プライベートサブネットのサーバーに外部から接続する際に、**パブリックサブネットに配置**する<br>
・VPCエンドポイント：AWSリソースを扱う際に、インターネットではなく**AWSネットワークを経由**してアクセスする機能<br>
・VPN / Direct Connect：パブリックサブネットに踏み台を配置せず、**セキュアにプライベートサブネットのサーバーにアクセス**する機能

セキュリティグループの絞り方には、以下の2つがある。

・接続元をIPアドレスベースで制御する方法<br>
・セキュリティIDベースで制御する方法
