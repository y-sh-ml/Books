## 1.5　カスタムクラス定義

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def draw(self):
    plt.plot(self.x, self.y, marker='o', markersize=10, c='k')
```

```python
p1 = Point(2, 3)
p2 = Point(-1, -2)

print(p1.x, p1.y)
print(p2.x, p2.y)
```

```python
plt.figure(figsize=(5, 5))
p1.draw()
p2.draw()
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.show()
```

```python
class Circle1(Point):
  def __init__(self, x, y, r):
    super().__init__(x, y)
    self.r = r
```

```python
c1_1 = Circle1(1, 0, 2)
print(c1_1.x, c1_1.y, c1_1.r)
```

```python
plt.figure(figsize=(5, 5))
ax = plt.subplot()
p1.draw()
p2.draw()
c1_1.draw()
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.show()
```

```python
class Circle2(Point):
  def __init__(self, x, y, r):
    super().__init__(x, y)
    self.r = r

  def draw(self):
    c = patches.Circle(xy=(self.x, self.y), radius=self.r, fc='b', ec='k')
    ax.add_patch(c)
```

```python
c2_1 = Circle2(1, 0, 2)
```

```python
plt.figure(figsize=(5, 5))
ax = plt.subplot()
p1.draw()
p2.draw()
c2_1.draw()
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.show()
```

```python
class Circle3(Point):
  def __init__(self, x, y, r):
    super().__init__(x, y)
    self.r = r

  def draw(self):
    super().draw()

    c = patches.Circle(xy=(self.x, self.y), radius=self.r, fc='b', ec='k')
    ax.add_patch(c)
```

```python
plt.figure(figsize=(5, 5))
c3_1 = Circle3(1, 0, 2)
ax = plt.subplot()
p1.draw()
p2.draw()
c3_1.draw()
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.show()
```
