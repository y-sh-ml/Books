## 1.6　インスタンス関数として使う

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class H:
  def __call__(self, x):
    return 2*x**2 + 2
```

```python
x = np.arange(-2, 2.1, 0.25)
print(x)

h = H()

y = h(x)
print(y)
```

```python
plt.figure(figsize=(5, 5))
plt.plot(x, y)
plt.show()
```
