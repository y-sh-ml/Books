## 1.4　「微分」と Python での「数値微分」の実装

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (2 * x**2 + 2)
```

```python
x = np.arange(-2, 2.1, 0.25)
```

```python
y = f(x)
print(y)
```

```python
def fdiff(f):

  def diff(x):
    h = 1e-6
    return (f(x+h) - f(x-h)) / (2*h)

  return diff
```

```python
diff = fdiff(f)
y_dash = diff(x)
print(y_dash)
```

```python
plt.figure(figsize=(5, 5))
plt.plot(x, y, label=r'y = f(x)', c='b')
plt.plot(x, y_dash, label=r"y = f '(x)", c='k')
plt.legend()
plt.show()
```

```python
def g(x):
  return 1 / (1+ np.exp(-x))
```

```python
y = g(x)
print(y)
```

```python
plt.figure(figsize=(5, 5))
plt.plot(x, y)
plt.show()
```

```python
diff = fdiff(g)
y_dash = diff(x)
print(y_dash)
```

```python
plt.figure(figsize=(5, 5))
plt.plot(x, y, label=r'y = f(x)', c = 'b')
plt.plot(x, y_dash, label=r"y = f '(x)", c='k')
plt.legend()
plt.show()
```
