## 1.2　コンテナデータ型の変数にご用心

```python
import torch
import numpy as np
```

```python
x = np.array([5, 7, 9])
y = x

print(x)
print(y)
```

```python
x[1] = -1

print(x)
print(y)
```

```python
x = np.array([5, 7, 9])
y = x.copy()
x[1] = -1

print(x)
print(y)
```

```python
x1 = torch.ones(5)
print(x1)

x2 = x1.data.numpy()
print(x2)
```

```python
x1[1] = -1

print(x1)
print(x2)
```

```python
x1 = torch.ones(5)
x2 = x1.data.numpy().copy()
x1[1] = -1

print(x1)
print(x2)
```
