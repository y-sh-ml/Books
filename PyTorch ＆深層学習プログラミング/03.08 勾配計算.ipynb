```python
import torch
import numpy as np
import japanize_matplotlib
import matplotlib.pyplot as plt
from torchviz import make_dot
from IPython.display import display
```

```python
%matplotlib inline
```

```python
sampleData1 = np.array([
    [166, 58.7],
    [176.0, 75.7],
    [171.0, 62.1],
    [173.0, 70.4],
    [169.0, 60.1]
])

x = sampleData1[:, 0]
y = sampleData1[:, 1]

X = x - x.mean()
Y = y - y.mean()

X = torch.tensor(X).float()
Y = torch.tensor(Y).float()

W = torch.tensor(1.0, requires_grad=True).float()
B = torch.tensor(1.0, requires_grad=True).float()

def pred(X):
  return W * X + B

Yp = pred(X)

def mse(Yp, Y):
  loss = ((Yp - Y) ** 2).mean()
  return loss

loss = mse(Yp, Y)
```

```python
loss.backward()
print(W.grad)
print(B.grad)
```
