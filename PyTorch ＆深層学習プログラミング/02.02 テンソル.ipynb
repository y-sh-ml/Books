## 2.2　テンソル

```python
import torch
import numpy as np
```

```python
r0 = torch.tensor(1.0).float()
print(type(r0))
print(r0.dtype)
```

```python
print(r0.shape)
print(r0.data)
```

```python
r1_np = np.array([1, 2, 3, 4, 5])
print(r1_np.shape)

r1 = torch.tensor(r1_np).float()

print(r1.dtype)
print(r1.shape)
print(r1.data)
```

```python
r2_np = np.array([[1, 5, 6], [4, 3, 2]])
print(r2_np.shape)

r2 = torch.tensor(r2_np).float()

print(r2.shape)
print(r2.data)
```

```python
torch.manual_seed(123)

r3 = torch.randn((3, 2, 2))
print(r3.shape)
print(r3.data)
```

```python
r4 = torch.ones((2, 3, 2, 2))
print(r4.shape)
print(r4.data)
```

```python
r5 = r1.long()
print(r5.dtype)
print(r5)
```

```python
r6 = r3.view(3, -1)
print(r6.shape)
print(r6.data)
```

```python
r7 = r3.view(-1)
print(r7.shape)
print(r7.data)
```

```python
print('requires_grad: ', r1.requires_grad)
print('device: ', r1.device)
```

```python
item = r0.item()
print(type(item))
print(item)
```
