## 3.5　データ前処理


```python
import torch
import numpy as np
import japanize_matplotlib
import matplotlib.pyplot as plt
from torchviz import make_dot
from IPython.display import display
```

```python
%matplotlib inline
```

```python
sampleData1 = np.array([
    [166, 58.7],
    [176.0, 75.7],
    [171.0, 62.1],
    [173.0, 70.4],
    [169.0, 60.1]
])

print(sampleData1)
```

```python
x = sampleData1[:, 0]
y = sampleData1[:, 1]
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(x, y, c='k', s=50)
plt.xlabel('$x$: 身長（cm）')
plt.ylabel('$y$: 体重（kg）')
plt.title('身長と体重の関係')
plt.show()
```

```python
X = x - x.mean()
Y = y - y.mean()
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(X, Y, c='k', s=50)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('加工後の身長と体重の関係')
plt.show()
```
