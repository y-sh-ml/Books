```python
X_np = np.arange(-4, 4.1, 0.25)
x = torch.tensor(X_np).float()
y = torch.sigmoid(x)

plt.figure(figsize=(5, 5))
plt.title('シグモイド関数のグラフ')
plt.plot(x.data, y.data)
plt.show()
```
