### 2.2.6　Numpy でのテンソル操作

```python
my_slice = train_images[10:100]
print(my_slice.shape)
```

```python
my_slice = train_images[10:100, :, :]
print(my_slice.shape)
my_slice = train_images[10:100, 0:28, 0:28]
print(my_slice.shape)
```

```python
my_slice = train_images[:, 14:, 14:]
print(my_slice.shape)
my_slice = train_images[:, 7:-7, 7:-7]
print(my_slice.shape)
```
